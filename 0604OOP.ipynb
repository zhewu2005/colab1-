我是一個開發人員，先幫我建立一個學生類別，該物件包含學生的學號、學生的姓名、學生的成績、學生的性別，用一個list來管理全班學生，幫我撰寫新增、修改、刪除和查詢全部學生的學號、學生的姓名、學生的成績、學生的性別，最後寫一個新增、修改、刪除、查詢全部學生和離開的使用者介面。

```python
# prompt: 我是一個開發人員，先幫我建立一個學生類別，該物件包含學生的學號、學生的姓名、學生的成績、學生的性別，用一個SQL list來管理全班學生，幫我撰寫新增、修改、刪除和查詢全部學生的學號、學生的姓名、學生的成績、學生的性別，最後寫一個新增、修改、刪除、查詢全部學生和離開的使用者介面

class Student:
    def __init__(self, student_id, name, score, gender):
        self.student_id = student_id
        self.name = name
        self.score = score
        self.gender = gender

    def __str__(self):
        return f"學號: {self.student_id}, 姓名: {self.name}, 成績: {self.score}, 性別: {self.gender}"

# Use a list to manage all students
students = []

def add_student():
    student_id = input("請輸入學生學號: ")
    name = input("請輸入學生姓名: ")
    score = input("請輸入學生分數: ")
    gender = input("請輸入學生性別: ")
    new_student = Student(student_id, name, score, gender)
    students.append(new_student)
    print("學生資料新增成功!")

def modify_student():
    student_id = input("請輸入要修改資料的學生學號: ")
    found_student = None
    for student in students:
        if student.student_id == student_id:
            found_student = student
            break

    if found_student:
        print(f"找到學生: {found_student}")
        found_student.name = input("請輸入新的學生姓名: ")
        found_student.score = input("請輸入新的學生分數: ")
        found_student.gender = input("請輸入新的學生性別: ")
        print("學生資料修改成功!")
    else:
        print("找不到該學號的學生.")

def delete_student():
    student_id = input("請輸入要刪除資料的學生學號: ")
    global students  # Declare students as global to modify the list
    initial_length = len(students)
    students = [student for student in students if student.student_id != student_id]
    if len(students) < initial_length:
        print("學生資料刪除成功!")
    else:
        print("找不到該學號的學生.")

def query_all_students():
    if not students:
        print("目前沒有學生資料.")
    else:
        print("--- 所有學生資料 ---")
        for student in students:
            print(student)
        print("--------------------")

def main():
    while True:
        print("\n--- 學生管理系統 ---")
        print("1. 新增學生")
        print("2. 修改學生")
        print("3. 刪除學生")
        print("4. 查詢所有學生")
        print("5. 離開")

        choice = input("請選擇功能 (1-5): ")

        if choice == '1':
            add_student()
        elif choice == '2':
            modify_student()
        elif choice == '3':
            delete_student()
        elif choice == '4':
            query_all_students()
        elif choice == '5':
            print("離開學生管理系統. 掰掰!")
            break
        else:
            print("無效的選擇，請重新輸入.")

if __name__ == "__main__":
    main()
```
